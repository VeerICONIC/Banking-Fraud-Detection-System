```python
import pandas as pd
import numpy as np
import joblib
```

```python
model = joblib.load("../models/best_model.pkl")

X_test = joblib.load("../models/X_test.pkl")

y_test = joblib.load("../models/y_test.pkl")
```

```python
fraud_probability = model.predict_proba(X_test)[:, 1]

predictions = model.predict(X_test)
```

```python
risk_df = X_test.copy()

risk_df["Actual"] = y_test.values

risk_df["Prediction"] = predictions

risk_df["Fraud_Probability"] = fraud_probability
```

```python
def assign_risk(prob):

    if prob < 0.20:
        return "Low"

    elif prob < 0.50:
        return "Medium"

    elif prob < 0.80:
        return "High"

    else:
        return "Critical"
```

```python
risk_df["Risk_Level"] = risk_df["Fraud_Probability"].apply(assign_risk)

risk_df.head()
```

```python
def recommend_action(level):

    if level == "Low":
        return "Approve"

    elif level == "Medium":
        return "Manual Review"

    elif level == "High":
        return "Hold Transaction"

    else:
        return "Block Transaction"
```

```python
risk_df["Recommended_Action"] = risk_df["Risk_Level"].apply(
    recommend_action
)

risk_df.head()
```

```python
risk_df["Confidence"] = np.where(

    risk_df["Prediction"] == 1,

    risk_df["Fraud_Probability"],

    1 - risk_df["Fraud_Probability"]

)
```

```python
def confidence_level(conf):

    if conf >= 0.95:
        return "Very High"

    elif conf >= 0.80:
        return "High"

    elif conf >= 0.60:
        return "Medium"

    else:
        return "Low"
```

```python
risk_df["Confidence_Level"] = risk_df["Confidence"].apply(
    confidence_level
)

risk_df.head()
```
