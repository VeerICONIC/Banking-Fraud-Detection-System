```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.max_columns",None)

plt.style.use("ggplot")
```

```python
df = pd.read_csv("../data/processed/banking_fraud.csv")

print(df.shape)
df.head()
```

```python
print("="*60)
print("Dataset Shape")
print(df.shape)

print("="*60)
print("Data Types")
display(df.dtypes.value_counts())

print("="*60)
print("Duplicate Rows")
print(df.duplicated().sum())
```

```python
missing = pd.DataFrame({
    "Missing Values":df.isnull().sum(),
    "Percentage":round(df.isnull().mean()*100,2)
})

missing = missing[missing["Missing Values"]>0]

missing = missing.sort_values(
    by="Percentage",
    ascending=False
)

missing.head(25)
```

```python
plt.figure(figsize=(12,8))

missing.head(20)["Percentage"].plot(kind="barh")

plt.title("Top 20 Features with Missing Values")

plt.xlabel("Missing Percentage")

plt.show()
```

```python
plt.figure(figsize=(6,6))

sns.countplot(
    x="isFraud",
    data=df
)

plt.title("Fraud vs Non-Fraud Transactions")

plt.show()
```

```python
fraud_percent = (
    df["isFraud"]
    .value_counts(normalize=True)
    *100
)

display(fraud_percent)
```

```python
df["TransactionAmt"].describe()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(
    df["TransactionAmt"],
    bins=100
)

plt.xlim(0,5000)

plt.title("Transaction Amount Distribution")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    x="isFraud",
    y="TransactionAmt",
    data=df
)

plt.ylim(0,5000)

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(
    np.log1p(df["TransactionAmt"]),
    bins=100
)

plt.title("Log Transaction Amount")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x="ProductCD",
    data=df
)

plt.show()
```

```python
product_fraud = (
    df
    .groupby("ProductCD")["isFraud"]
    .mean()
    .sort_values(ascending=False)
)

product_fraud
```

```python
plt.figure(figsize=(8,5))

product_fraud.plot(
    kind="bar",
    color="tomato"
)

plt.ylabel("Fraud Rate")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    x="card4",
    data=df
)

plt.show()
```

```python
card_fraud = (
    df
    .groupby("card4")["isFraud"]
    .mean()
)

card_fraud.plot(kind="bar")

plt.ylabel("Fraud Rate")

plt.show()
```

```python
top_email = df["P_emaildomain"].value_counts().head(15)

plt.figure(figsize=(12,5))

top_email.plot(kind="bar")

plt.title("Top Email Domains")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x="DeviceType",
    data=df
)

plt.show()
```

```python
device_rate = (
    df
    .groupby("DeviceType")["isFraud"]
    .mean()
)

device_rate.plot(kind="bar")

plt.ylabel("Fraud Rate")

plt.show()
```

```python
hours = (
    df["TransactionDT"]
    /(60*60)
)%24

plt.figure(figsize=(12,5))

sns.histplot(hours,bins=24)

plt.title("Transactions by Hour")

plt.show()
```

```python
hour_df = pd.DataFrame({
    "Hour":hours,
    "Fraud":df["isFraud"]
})

hour_rate = (
    hour_df
    .groupby("Hour")["Fraud"]
    .mean()
)

plt.figure(figsize=(12,5))

hour_rate.plot()

plt.ylabel("Fraud Rate")

plt.show()
```

```python
numeric = df.select_dtypes(include=np.number)

corr = numeric.corr()["isFraud"].abs()

corr = corr.sort_values(
    ascending=False
)

corr.head(20)
```

```python
top = corr.head(15)

plt.figure(figsize=(10,6))

top.plot(kind="barh")

plt.show()
```

```python
cols = corr.head(10).index

plt.figure(figsize=(10,8))

sns.heatmap(
    df[cols].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.show()
```
